# Titanic Dataset

```python
import os
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader, random_split
```

```python
# Titanic 데이터를 위한 PyTorch Dataset 클래스
# 특징과 레이블을 텐서로 저장하고 데이터 크기와 형태 정의
class TitanicDataset(Dataset):
  def __init__(self, X, y):
    self.X = torch.FloatTensor(X)
    self.y = torch.LongTensor(y)

  def __len__(self):
    return len(self.X)

  def __getitem__(self, idx):
    feature = self.X[idx]
    target = self.y[idx]
    return {'input': feature, 'target': target}

  def __str__(self):
    str = "Data Size: {0}, Input Shape: {1}, Target Shape: {2}".format(
      len(self.X), self.X.shape, self.y.shape
    )
    return str
```

```python
# 테스트용 Titanic 데이터를 위한 PyTorch Dataset 클래스
# 특징을 텐서로 저장하고 데이터 크기와 형태 정의
class TitanicTestDataset(Dataset):
  def __init__(self, X):
    self.X = torch.FloatTensor(X)

  def __len__(self):
    return len(self.X)

  def __getitem__(self, idx):
    feature = self.X[idx]
    return {'input': feature}

  def __str__(self):
    str = "Data Size: {0}, Input Shape: {1}".format(
      len(self.X), self.X.shape
    )
    return str
```

```python
# Titanic 데이터셋을 불러와 전처리하고
# 학습·검증·테스트용 PyTorch Dataset 객체로 분할하는 함수
def get_preprocessed_dataset():
    CURRENT_FILE_PATH = os.path.dirname(os.path.abspath(os.path.join('homework2')))
    print(CURRENT_FILE_PATH)

    train_data_path = os.path.join(CURRENT_FILE_PATH, "train.csv")
    test_data_path = os.path.join(CURRENT_FILE_PATH, "test.csv")

    train_df = pd.read_csv(train_data_path)
    test_df = pd.read_csv(test_data_path)

    all_df = pd.concat([train_df, test_df], sort=False)

    all_df = get_preprocessed_dataset_1(all_df)

    all_df = get_preprocessed_dataset_2(all_df)

    all_df = get_preprocessed_dataset_3(all_df)

    all_df = get_preprocessed_dataset_4(all_df)

    all_df = get_preprocessed_dataset_5(all_df)

    all_df = get_preprocessed_dataset_6(all_df)

    train_X = all_df[~all_df["Survived"].isnull()].drop("Survived", axis=1).reset_index(drop=True)
    train_y = train_df["Survived"]

    test_X = all_df[all_df["Survived"].isnull()].drop("Survived", axis=1).reset_index(drop=True)

    dataset = TitanicDataset(train_X.values, train_y.values)
    #print(dataset)
    train_dataset, validation_dataset = random_split(dataset, [0.8, 0.2])
    test_dataset = TitanicTestDataset(test_X.values)
    #print(test_dataset)

    return train_dataset, validation_dataset, test_dataset
```

```python
# Pclass별 평균 Fare 값을 이용해 Fare 결측치를 채우는 함수
def get_preprocessed_dataset_1(all_df):
    # Pclass별 Fare 평균값을 사용하여 Fare 결측치 메우기
    Fare_mean = all_df[["Pclass", "Fare"]].groupby("Pclass").mean().reset_index()
    Fare_mean.columns = ["Pclass", "Fare_mean"]
    all_df = pd.merge(all_df, Fare_mean, on="Pclass", how="left")
    all_df.loc[(all_df["Fare"].isnull()), "Fare"] = all_df["Fare_mean"]

    return all_df
```

```python
# 이름 컬럼을 '성', '경칭', '이름'으로 분리하여 all_df에 추가하는 함수
def get_preprocessed_dataset_2(all_df):
    # name을 세 개의 컬럼으로 분리하여 다시 all_df에 합침
    name_df = all_df["Name"].str.split("[,.]", n=2, expand=True)
    name_df.columns = ["family_name", "honorific", "name"]
    name_df["family_name"] = name_df["family_name"].str.strip()
    name_df["honorific"] = name_df["honorific"].str.strip()
    name_df["name"] = name_df["name"].str.strip()
    all_df = pd.concat([all_df, name_df], axis=1)

    return all_df
```

```python
# 경칭(honorific)별 나이(Age) 평균값을 사용해 Age 결측치를 채우는 함수
def get_preprocessed_dataset_3(all_df):
    # honorific별 Age 평균값을 사용하여 Age 결측치 메우기
    honorific_age_mean = all_df[["honorific", "Age"]].groupby("honorific").median().round().reset_index()
    honorific_age_mean.columns = ["honorific", "honorific_age_mean", ]
    all_df = pd.merge(all_df, honorific_age_mean, on="honorific", how="left")
    all_df.loc[(all_df["Age"].isnull()), "Age"] = all_df["honorific_age_mean"]
    all_df = all_df.drop(["honorific_age_mean"], axis=1)

    return all_df
```

```python
# 가족 수와 혼자 탑승 여부를 나타내는 컬럼을 추가하고
# 학습에 불필요한 컬럼을 제거하는 함수
def get_preprocessed_dataset_4(all_df):
    # 가족수(family_num) 컬럼 새롭게 추가
    all_df["family_num"] = all_df["Parch"] + all_df["SibSp"]

    # 혼자탑승(alone) 컬럼 새롭게 추가
    all_df.loc[all_df["family_num"] == 0, "alone"] = 1
    all_df["alone"].fillna(0, inplace=True)

    # 학습에 불필요한 컬럼 제거
    all_df = all_df.drop(["PassengerId", "Name", "family_name", "name", "Ticket", "Cabin"], axis=1)

    return all_df
```

```python
# honorific 값을 주요 범주(Mr, Miss, Mrs, Master)와 기타(other)로 줄이고, Embarked의 결측치를 "missing"으로 채우는 함수
def get_preprocessed_dataset_5(all_df):
    # honorific 값 개수 줄이기
    all_df.loc[
    ~(
            (all_df["honorific"] == "Mr") |
            (all_df["honorific"] == "Miss") |
            (all_df["honorific"] == "Mrs") |
            (all_df["honorific"] == "Master")
    ),
    "honorific"
    ] = "other"
    all_df["Embarked"].fillna("missing", inplace=True)

    return all_df
```

```python
# 카테고리 변수들을 LabelEncoder로 수치형으로 변환하는 함수
def get_preprocessed_dataset_6(all_df):
    # 카테고리 변수를 LabelEncoder를 사용하여 수치값으로 변경하기
    category_features = all_df.columns[all_df.dtypes == "object"]
    from sklearn.preprocessing import LabelEncoder
    for category_feature in category_features:
        le = LabelEncoder()
        if all_df[category_feature].dtypes == "object":
          le = le.fit(all_df[category_feature])
          all_df[category_feature] = le.transform(all_df[category_feature])

    return all_df
```

```python
# 입력 크기와 출력 크기를 설정해 3개의 선형 계층과 활성화 함수로 구성된 신경망 모델 클래스
from torch import nn
class MyModel_1(nn.Module):
  def __init__(self, n_input, n_output):
    super().__init__()

    self.model = nn.Sequential(
      nn.Linear(n_input, 30),
      nn.ELU(),
      nn.Linear(30, 30),
      nn.ELU(),
      nn.Linear(30, n_output),
    )

  def forward(self, x):
    x = self.model(x)
    return x
```

```python
# 테스트 데이터 로더에서 배치를 가져와 모델 예측을 출력하는 함수
# 각 샘플의 예측 결과를 인덱스와 함께 표시
def test(test_data_loader):
  print("[TEST]")
  batch = next(iter(test_data_loader))
  print("{0}".format(batch['input'].shape))
  my_model = MyModel_1(n_input=11, n_output=1)
  output_batch = my_model(batch['input'])
  prediction_batch = torch.argmax(output_batch, dim=1)
  for idx, prediction in enumerate(prediction_batch, start=892):
      print(idx, prediction.item())
```

```python
# 학습, 검증, 테스트 데이터셋을 생성하고, 각 데이터셋 크기를 출력하며 데이터 로더를 설정하는 메인 실행 블록
if __name__ == "__main__":
  train_dataset, validation_dataset, test_dataset = get_preprocessed_dataset()

  print("train_dataset: {0}, validation_dataset.shape: {1}, test_dataset: {2}".format(
    len(train_dataset), len(validation_dataset), len(test_dataset)
  ))
  print("#" * 50, 1)

  for idx, sample in enumerate(train_dataset):
    print("{0} - {1}: {2}".format(idx, sample['input'], sample['target']))

  print("#" * 50, 2)

  train_data_loader = DataLoader(dataset=train_dataset, batch_size=16, shuffle=True)
  validation_data_loader = DataLoader(dataset=validation_dataset, batch_size=16, shuffle=True)
  test_data_loader = DataLoader(dataset=test_dataset, batch_size=len(test_dataset))
```

```python
# 학습 및 검증 데이터 로더에서 각 배치의 입력과 타겟 형상을 출력하고, 테스트 데이터에서 예측 결과를 출력
print("[TRAIN]")
for idx, batch in enumerate(train_data_loader):
    print("{0} - {1}: {2}".format(idx, batch['input'].shape, batch['target'].shape))

print("[VALIDATION]")
for idx, batch in enumerate(validation_data_loader):
    print("{0} - {1}: {2}".format(idx, batch['input'].shape, batch['target'].shape))

print("#" * 50, 3)

test(test_data_loader)
```

```python
import torch
import csv
from torch import nn, optim
from torch.utils.data import random_split, DataLoader
from datetime import datetime
import wandb
import argparse

from pathlib import Path
BASE_PATH = str(Path(os.path.join('homework2')).resolve().parent.parent.parent) # BASE_PATH: /Users/yhhan/git/link_dl
print(BASE_PATH, "!!!!!!!")

import sys
sys.path.append(BASE_PATH)


def get_data():
  train_dataset, validation_dataset, _ = get_preprocessed_dataset()
  train_data_loader = DataLoader(dataset=train_dataset, batch_size=16, shuffle=True)
  validation_data_loader = DataLoader(dataset=validation_dataset, batch_size=16, shuffle=True)

  return train_data_loader, validation_data_loader


# 두 개의 은닉층을 가지는 신경망 모델 클래스, wandb 설정에 따라 입력층과 은닉층 크기를 조정
class MyModel(nn.Module):
  def __init__(self, n_input, n_output):
    super().__init__()

    self.model = nn.Sequential(
      nn.Linear(n_input, wandb.config.n_hidden_unit_list[0]),
      nn.ELU(),
      nn.Linear(wandb.config.n_hidden_unit_list[0], wandb.config.n_hidden_unit_list[1]),
      nn.ELU(),
      nn.Linear(wandb.config.n_hidden_unit_list[1], n_output),
    )

  def forward(self, x):
    x = self.model(x)
    return x



# 모델과 SGD 옵티마이저를 설정하는 함수
def get_model_and_optimizer():
  my_model = MyModel(n_input=11, n_output=2)
  optimizer = optim.SGD(my_model.parameters(), lr=wandb.config.learning_rate)

  return my_model, optimizer


# 지정된 에포크 동안 MSE 손실로 모델을 학습하고, 매 에포크별로 wandb에 학습 및 검증 손실을 기록
def training_loop(model, optimizer, train_data_loader, validation_data_loader):
  n_epochs = wandb.config.epochs
  loss_fn = nn.CrossEntropyLoss()  # Use a built-in loss function
  next_print_epoch = 100

  for epoch in range(1, n_epochs + 1):
    loss_train = 0.0
    num_trains = 0
    for train_batch in train_data_loader:
      input = train_batch['input']   # train_batch에서 딕셔너리 형태로 'input'과 'target'을 가져옴
      target = train_batch['target']
      output_train = model(input)
      loss = loss_fn(output_train, target)
      loss_train += loss.item()
      num_trains += 1

      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

    loss_validation = 0.0
    num_validations = 0
    with torch.no_grad():
      for validation_batch in validation_data_loader:
        input = validation_batch['input']   # validation_batch에서도 'input'과 'target'을 가져옴
        target = validation_batch['target']
        output_validation = model(input)
        loss = loss_fn(output_validation, target)
        loss_validation += loss.item()
        num_validations += 1

    wandb.log({
      "Epoch": epoch,
      "Training loss": loss_train / num_trains,
      "Validation loss": loss_validation / num_validations
    })

    if epoch >= next_print_epoch:
      print(
        f"Epoch {epoch}, "
        f"Training loss {loss_train / num_trains:.4f}, "
        f"Validation loss {loss_validation / num_validations:.4f}"
      )
      next_print_epoch += 100

def create_submission(model, test_data_loader, output_file="submission.csv"):
    # 모델을 평가 모드로 전환 (dropout, batch norm 등 적용 안함)
    model.eval()
    
    # 예측을 저장할 리스트
    predictions = []
    
    # 테스트 데이터를 배치 단위로 예측
    with torch.no_grad():
        for batch in test_data_loader:
            input_data = batch['input']
            output = model(input_data)  # 모델로 예측
            pred = torch.argmax(output, dim=1)  # 가장 높은 확률의 클래스 선택
            predictions.extend(pred.tolist())  # 예측 결과를 리스트에 저장
    
    # CSV 파일 작성
    with open(output_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["PassengerId", "Survived"])  # CSV 파일 헤더 작성
        
        # 테스트 데이터의 PassengerId와 예측 결과 매칭하여 저장 (PassengerId는 892부터 시작)
        for idx, prediction in enumerate(predictions, start=892):
            writer.writerow([idx, prediction])
    
    print(f"Submission file {output_file} has been created.")


# 실행 시 argparse로 설정값을 받아 wandb를 초기화하고, 데이터 로딩 및 학습 루프를 실행
def main(args):
  current_time_str = datetime.now().astimezone().strftime('%Y-%m-%d_%H-%M-%S')

  config = {
    'epochs': args.epochs,
    'batch_size': args.batch_size,
    'learning_rate': 1e-3,
    'n_hidden_unit_list': [20, 20],
  }

  wandb.init(
    # mode="online" if args.wandb else "disabled",
    project="titanicDataset",
    notes="My first wandb experiment",
    tags=["my_model", "titanic"],
    # name=current_time_str,
    config=config
  )
  print(args)
  print(wandb.config)

  train_data_loader, validation_data_loader = get_data()

  linear_model, optimizer = get_model_and_optimizer()

  print("#" * 50, 1)

  training_loop(
    model=linear_model,
    optimizer=optimizer,
    train_data_loader=train_data_loader,
    validation_data_loader=validation_data_loader
  )

  # 테스트 데이터 로더 생성 (get_preprocessed_dataset에서 test_dataset 불러오기)
  _, _, test_dataset = get_preprocessed_dataset()
  test_data_loader = DataLoader(dataset=test_dataset, batch_size=len(test_dataset), shuffle=False)

  # 훈련 완료 후 테스트 데이터 예측 및 submission.csv 생성
  create_submission(model=linear_model, test_data_loader=test_data_loader, output_file="submission.csv")
  wandb.finish()


# https://docs.wandb.ai/guides/track/config
if __name__ == "__main__":
  parser = argparse.ArgumentParser()

  parser.add_argument(
    "--wandb", action=argparse.BooleanOptionalAction, default=False, help="True or False"
  )

  parser.add_argument(
    "-b", "--batch_size", type=int, default=512, help="Batch size (int, default: 512)"
  )

  parser.add_argument(
    "-e", "--epochs", type=int, default=1_000, help="Number of training epochs (int, default:1_000)"
  )

  args, unknown = parser.parse_known_args()

  main(args)
```

### ReLU, LeakyReLU, PReLU, ELU, Sigmoid, Tanh의 Training Loss, Validation Loss 그래프
https://wandb.ai/gwak1016-korea-university-of-technology-and-education/titanicDataset/workspace

ReLU, LeakyReLU, PReLU, ELU, Sigmoid, Tanh 활성화 함수를 사용해봤다. 
먼저 **ReLU**의 Training Loss는 안정적으로 감소하며 낮은 값에 수렴 중이고, Validation Loss는 일정하지만 변동성이 크지 않고 낮은 수준을 유지한다. 따라서 안정적이고 좋은 성능을 갖는다고 할 수 있다.
**LeakyReLU**의 Training Loss는 감소하는 추세이지만, 다른 함수들에 비해 높게 유지된다. 그리고, Validation Loss는 큰 변동성이 있으며, 불안정하게 유지된다. 따라서 불안정하며 성능이 떨어진다.
**PReLU**의 Training Loss는 안정적으로 감소하고, 낮은 값 유지한다. 그리고 Validation Loss는 비교적 안정적이며 낮은 수준을 유지한다. 따라서 안정적이고 성능이 우수하다.
**ELU**의 Training Loss는 매우 안정적이고 빠르게 감소하고, validation Loss는 변동성이 적고 낮은 값을 유지한다. 따라서 매우 안정적인 성능을 보인다.
**Sigmoid**의 Training Loss는 감소가 느리고 비교적 높게 유지된다. 그리고 Validation Loss는 높은 변동성을 보이며 높은 값에서 유지된다. 따라서 성능이 떨어진다.
**Tanh**의 Training Loss는 감소 속도가 느리고 높게 유지하고, Validation Loss는 높은 변동성을 보이며 성능이 불안정하다 따라서 성능이 낮다.
위의 결과로 보았을 때 ELU가 가장 좋은 성능을 보인다고 할 수 있다.
